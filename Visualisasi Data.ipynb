# Import dataset
Memuat dataset dan kemudian menampilkan data tersebut agar bisa diproses

```python
data = pd.read_csv("Flipkart.csv")
```

# Menampilkan 5 data teratas pada dataset

```python
data.head()
```

# Menampilkan ringkasan dari dataset

```python
data.info()
```

# Menampilkan data unik pada kolom Memory

```python
data['Memory'].unique()
```

# Menampilkan data berdasarkan dengan data yang telah ditentukan pada variable static

```python
Memory_Product = '8 GB'
data[data['Memory']==Memory_Product]
```

# Menampung dan menampilkan data brand yang unik berdasarkan data selling price yang bernilai lebih dari 1000

```python
unique_Brand = data[data.Selling_Price>1000]['Brand'].unique()
unique_Brand.sort()
unique_Brand
```

# Menampilkan jumlah harga penjualan dari brand yang berdasarkan memory yang telah ditentukan

```python
Brand_Selling_Price_case = []
for i in unique_Brand:
    Brand_Selling_Price_case.append(data[data.Selling_Price>1000][data['Brand']==i][data['Memory']==Memory_Product].Selling_Price.sum())
```

```python
print(set(zip(unique_Brand, Brand_Selling_Price_case)))
```

# Menampung dan menampilkan data Model yang unik berdasarkan data selling price yang bernilai lebih dari 1000

```python
unique_model = data['Model'][data.Selling_Price>1000].unique()
unique_model
```

```python
Model_Selling_Price_case = []
for i in unique_model:
    Model_Selling_Price_case.append(data[data.Selling_Price>1000][data['Model']==i][data['Memory']==Memory_Product].Selling_Price.sum())
```

```python
print(set(zip(unique_model, Model_Selling_Price_case)))
```

# Mengolah data yang terdapat NaN

Mengecek data yang terdapat NaN pada Model jika terindikasi maka data tersebut akan ditampung pada variable nan_indices.

```python
nan_indices = []
for i in range(len(unique_model)):
    if type(unique_model[i])==float:
        nan_indices.append(i)
```

Kemudian kita buat list untuk unique_model dan juga Model_Selling_Price_case.

```python
unique_model = list(unique_model)
Model_Selling_Price_case = list(Model_Selling_Price_case)
```

Kemudian kita hapus data yang NaN pada unique_model dan Model_Selling_Price_case dan kemudian menampilkan data unique_model yang telah berhasil diolah / bersih

```python
for i in nan_indices:
    unique_model.pop(i)
    Model_Selling_Price_case.pop(i)
    
unique_model
```

# Membuat Grafik dengan menggunakan Library matplotlib.pyplot 

Menampilkan grafik dari data unique_Brand dan Brand_Selling_Price_case

```python
plt.barh(unique_Brand, Brand_Selling_Price_case)
plt.show()
```

Menampilkan grafik dari data unique_model dan model_Selling_Price_case

```python
plt.figure(figsize=(50,40))
plt.barh(unique_model, Model_Selling_Price_case)
plt.show()
```

Menampilkan dengan Diagram Pie dari data unique_Brand dan Brand_Selling_Price_case

```python
plt.figure(figsize=(10,10))
plt.pie(Brand_Selling_Price_case)
plt.legend(unique_Brand)
plt.show()
```

Menampilkan dengan Diagram Pie dari data unique_model dan model_Selling_Price_case

```python
plt.figure(figsize=(20,20))
plt.pie(Model_Selling_Price_case)
plt.legend(unique_model, loc='best')
plt.show()
```
